Documentação: pandas.DataFrame.plot

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html?highlight=pandas%20plot#pandas.DataFrame.plot



```
dados.plot(x="Unidade da Federação", y="2008/Ago")
```

```
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
```

https://matplotlib.org/3.3.1/gallery/ticks_and_spines/tick-formatters.html

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.title("Valor por Unidade da Federação")
plt.show()
```
